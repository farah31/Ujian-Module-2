```python
#IQR, Mean, Median, Std, LowBound, UppBound dari Angka Harapan Hidup Benua Asia
dfAsia = dfCountry[dfCountry['Continent'] == 'Asia']
dfAsia

Q1 = dfAsia.LifeExpectancy.quantile(.25)
Q3 = dfAsia.LifeExpectancy.quantile(.75)
IQR = Q3 - Q1
LowBound = Q1 - (1.5*IQR)
UppBound = Q3 + (1.5*IQR)
median = dfAsia.LifeExpectancy.median()
mean = dfAsia.LifeExpectancy.mean()
std = dfAsia.LifeExpectancy.std()

print(f'Q1: {Q1}')
print(f'Q3: {Q3}')
print(f'IQR: {IQR}')
print(f'Batas Minimum: {LowBound}')
print(f'Batas Maksimum: {UppBound}')
print(f'Mean: {mean}')
print(f'Median: {median}')
print(f'Standar Deviasi: {std}')
```

```python
# Distrib Plot Life Expectancy Benua Asia
sns.distplot(dfAsia.LifeExpectancy)
plt.axvline(dfAsia.LifeExpectancy.mean(), color = 'g', label = 'mean')
plt.axvline(dfAsia.LifeExpectancy.median(), color = 'r', label = 'median')
plt.axvline(LowBound, color = 'black', label='Lower Bound')
plt.axvline(UppBound,color = 'yellow', label='Upper Bound')
plt.legend(loc = 'upper left')
```

```python
#Negara yang menjadi outlier angka harapan hidup
dfOutliermin = dfAsia[dfAsia['LifeExpectancy'] < LowBound]
dfOutlierminN = dfOutliermin['Name']
dfOutliermax = dfAsia[dfAsia['LifeExpectancy'] > UppBound]
dfOutliermaxN = dfOutliermax['Name']

print(f'Negara yang berada di bawah batas minimum angka harapan hidup: {dfOutlierminN}')
print(f'Negara yang berada di atas batas maksimum angka harapan hidup: {dfOutliermaxN}')
```
